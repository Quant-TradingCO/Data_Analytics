# **CORRELATION**

## **¿How to calculate correlations from financial returns?**

<font size="3"> In statistics, correlation or dependence is any statistical relationship, whether causal or not, between two random variables or bivariate data. Although in the broadest sense, "correlation" may indicate any type of association, in statistics it usually refers to the degree to which a pair of variables are linearly related. Here we we learn how to calculate linear correlation between financial returns.
<br><br>

```python
import warnings
warnings.filterwarnings('ignore')

import yfinance as yf  #Yahoo Finance
import datetime
import pandas as pd
import numpy as np
```

## **Get an specific ticker**

<font size="3"> Stocks have an specific identification code. Here we show how to download the prices history for what are known as the magnificent 7 stocks for an specific time frame. remember that we show how to download data from the yahoo finance api in the following link: <a href="https://quant-trading.co/how-to-download-data-from-yahoo-finance-api/" target="_blank">YAHOO FINANCE API</a>
<br><br>

```python
#ASSET
tickers_list = ['AAPL', 'GOOG', 'AMZN', 'TSLA', 'NVDA', 'MSFT','META']

multiple_prices = yf.download(tickers_list,period='5y')['Adj Close']
multiple_prices
```

<font size="3"> Now we can calculate the returns for the different tickers as shown below. We use a loop to do that:
<br><br>

```python
for elem in tickers_list:    
    multiple_prices['Return '+ elem] = multiple_prices[elem]/multiple_prices[elem].shift(1) - 1
    multiple_prices['Return '+ elem].iloc[0] = 0
    
multiple_prices
```

## **Calculate the correlation between 2 stocks**

<font size="3"> To calculate the correlation between 2 stocks we can use the numpy function corrcoef as shown below:
<br><br>

```python
np.corrcoef(multiple_prices['Return AAPL'],multiple_prices['Return GOOG'])
```

## **Calculate the correlation matrix of the returns**

<font size="3"> Since we have 7 stocks in this DataFrame, we would need to calculate 6 + 5 + 4 + 3 + 2 + 1 = 21 pairs of correlations between all stocks. Instead, we can calculate a correlation matrix for that purpose. The first thing we need to do is to filter the DataFrame to just leave the returns of the stocks.
<br><br>

```python
multiple_returns = multiple_prices[['Return AAPL','Return GOOG','Return AMZN','Return TSLA','Return NVDA','Return MSFT','Return META']]
multiple_returns
```

<font size="3"> After that, we applied the corr() function from the pandas library on the entire DataFrame to get the correlation matrix as shown below:
<br><br>

```python
corr_matrix = multiple_returns.corr()
corr_matrix
```

## **How do you know which stock has the highest correlation with another?**

<font size="3"> You can sort the values for an specific column to get the highest to lowest correlation among stock pairs. For example, in this case we take the AAPL return with the other magnificent 7 stocks. We note that the highest correlation is with the MSFT stock at 75.6732%, while the lowest is with TSLA stock at 48.7754%
<br><br>

```python
corr_matrix["Return AAPL"].sort_values(ascending=False)
```

## **Can we calculate correlation for different assets?**

<font size="3"> When we calculate correlations for a group of similar stocks, we usually get very high value for them. You say, that they have a very high correlation or that are highly correlated. In finance, when people talk about diversification, they usually like to have asset's with low (close to zero) or negative correlations. Let's explore the following ETFs and see their correlation. 
<br><br>
    
<font size="3"> They are:
<br><br>
    
<font size="3"> 
SPY -> S&P500 <br><br>
TLT -> Long Term Treasuries <br><br>
IAU -> Gold <br><br>
EMB -> Emerging Market Bonds <br><br>
USO -> Crude Oil <br><br>
EEM -> Emerging Market Stocks <br><br>
SHV -> Short Term Treasuries <br><br>

```python
#ASSET
tickers_list2 = ['SPY', 'TLT', 'IAU', 'EMB', 'USO', 'EEM','SHV']

multiple_prices2 = yf.download(tickers_list2,period='5y')['Adj Close']


for elem in tickers_list2:    
    multiple_prices2['Return '+ elem] = multiple_prices2[elem]/multiple_prices2[elem].shift(1) - 1
    multiple_prices2['Return '+ elem].iloc[0] = 0
    
multiple_returns2 = multiple_prices2[['Return SPY','Return TLT','Return IAU','Return EMB','Return USO','Return EEM','Return SHV']]
multiple_returns2.corr()
```

## **Rolling correlation**

<font size="3"> Correlation does vary with time. Sometimes is good to track how it has evolved trough different periods. Below you can see how to calculate rolling correlation for a pair of asset's returns:
<br><br>

```python
multiple_prices['Return AAPL'].rolling(window=180).corr(multiple_prices['Return GOOG'])
```
